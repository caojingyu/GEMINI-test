```python
#import pakages
import pandas as pd
import numpy as np
```

```python
#1. read files
admin_site1 = pd.read_csv('administrative_site1.csv')
admin_site2 = pd.read_csv('administrative_site2.csv')
lab = pd.read_csv('lab.csv')
display(admin_site1.head(), admin_site2.head(), lab.head())
```

```python
#join site1 & 2 as admin
admin_site2.columns = admin_site1.columns
admin = admin_site1.append(admin_site2, sort=False)
admin.head()
```

```python
admin.info()
```

```python
#2 length of stay
dtcols = ['admission_date', 'discharge_date', 'admission_time', 'discharge_time']
for dtcol in dtcols:
    admin[dtcol] = pd.to_datetime(admin[dtcol])
admin['length_of_stay'] = admin['discharge_date'] - admin['admission_date']
admin['length_of_stay'] = admin['length_of_stay'].dt.days
```

```python
#for each hospital
admin.groupby('hospital_ID')['length_of_stay'].mean()
```

```python
#3. combine admin & lab using ID
adm_cp = pd.merge(admin, lab, how='outer', on='ID')
adm_cp.head()
```

```python
#5.a. transform lab long to wide
lab_wide = lab.pivot_table(index='ID', columns='test_name', values='result_value', aggfunc='min')
lab_wide.head()
```

```python
lab_wide.info()
```

```python
#5.b. mean for urea plasma
lab_wide['Urea plasma'].dropna().mean()
```

```python
#6.a. remove NA in admin
adm_filtered = admin.dropna(axis='rows', thresh=2)
adm_filtered.head()
```

```python
#6.b. missing rate of adm_filtered
adm_filtered.isnull().sum()/adm_filtered.count()
```

```python
#7. copy lab
lab_impute = lab.copy()
lab_impute['result_date'] = pd.to_datetime(lab_impute['result_date'])
lab2002 = lab_impute[lab_impute['result_date'].dt.year == 2002]
lab2002['result_value'] = np.nan
lab2002.head()
```

```python
labrest = lab_impute[lab_impute['result_date'].dt.year != 2002]
restmean = labrest.groupby('test_name')['result_value'].mean()
for testname in restmean.keys():
    lab2002.loc[lab2002['test_name']==testname, 'result_value'] = restmean[testname]
lab_impute[lab_impute['result_date'].dt.year == 2002] = lab2002
lab_impute['result_value'].mean()
```

```python
#9. function year output
def year_output(year, output_file_name):
    adminyear = admin[admin['discharge_date'].dt.year==year]
    adminyear.to_csv(output_file_name)
```

```python
year_output(2005, 'Discharge_2005.csv')
```

```python
#10. fucntion compute hospital mean
def compute_hospital_mean(test_code, hospital_ID):
    data = pd.merge(admin, lab, how='outer', on='ID')
    datamean = data[(data['test_code']==test_code) & (data['hospital_ID']==hospital_ID)]['result_value'].mean()
    
    return datamean
    
```

```python
compute_hospital_mean('KPL', "St. Michael's Hospital")
```
